# 94.Nested while Loop in Python


Your code is using nested while loops to iterate over elements of two lists, `list1` and `list2`, and print pairs of elements from both lists. However, there's a logical issue in the nested loops. You're not resetting the value of `j` to 0 after the inner loop completes its execution. As a result, the inner loop executes only once because `j` is already equal to `len(list2)` after the first iteration of the outer loop. Here's the corrected code:

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
i = 0
# Iterate over elements of list1
while i < len(list1):
    j = 0  # Reset j to 0 for each iteration of the outer loop
    # Iterate over elements of list2
    while j < len(list2):
        # Print pairs of elements from list1 and list2
        print(list1[i], list2[j])
        j += 1
    print()
    i += 1
```

Now, the code will correctly print all possible pairs of elements from `list1` and `list2`.
