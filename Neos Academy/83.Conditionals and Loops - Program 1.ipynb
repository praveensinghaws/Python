# 83.Conditionals and Loops - Program 1


Your code correctly calculates the sum of numbers from `n` down to `1`. Here's the code with comments:

```python
# Ask the user to enter the value of n and convert it to an integer
n = int(input("Enter the value of n: "))

# Initialize the variable sum to store the sum of numbers
sum = 0

# Continue looping while n is greater than 0
while n > 0:
    # Add the current value of n to the sum
    sum += n  # Equivalent to: sum = sum + n
    # Decrement the value of n by 1 for the next iteration
    n -= 1   # Equivalent to: n = n - 1

# Print the sum of numbers
print(f"Sum is {sum}")
```

In this code:
- `n = int(input("Enter the value of n: "))` prompts the user to enter a value for `n` and converts the input to an integer.
- Inside the `while` loop, the current value of `n` is added to the `sum`.
- The value of `n` is then decremented by `1` in each iteration to progress towards the termination condition `n > 0`.
- Once `n` becomes `0`, the loop terminates, and the sum of numbers from the user input down to `1` is printed.
