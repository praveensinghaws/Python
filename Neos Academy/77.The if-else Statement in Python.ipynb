# 77.The if-else Statement in Python


This code snippet checks the value of the variable `age`. If the age is less than 18, it prints "You can't vote!". Otherwise, it prints "You can vote!". Finally, it prints "Done!" regardless of the age. Here's the code with comments:

```python
# Assign the value 20 to the variable age
age = 20

# Check if age is less than 18
if age < 18:
    # If age is less than 18, print the message
    print("You can't vote!")
else:
    # If age is not less than 18 (i.e., 18 or older), print the message
    print("You can vote!")

# Print "Done!" after the if-else block
print("Done!")
```

In this code:
- `age = 20` assigns the value `20` to the variable `age`.
- The `if` statement checks if the value of `age` is less than `18`.
- If the condition `age < 18` evaluates to `True`, the message "You can't vote!" is printed.
- If the condition is `False` (i.e., age is 18 or older), the `else` block is executed, and the message "You can vote!" is printed.
- Finally, the message "Done!" is printed regardless of the condition.

This code snippet achieves the same outcome as the previous one but uses a conditional expression (also known as a ternary operator) instead of an `if-else` statement. Here's the code with comments:

```python
# Assign the value 20 to the variable age
age = 20

# Use a conditional expression to print a message based on the value of age
# If age is less than 18, print "You can't vote!", otherwise print "You can vote!"
print("You can't vote!") if age < 18 else print("You can vote!")

# Print "Done!" after the conditional expression
print('Done!')
```

In this code:
- `age = 20` assigns the value `20` to the variable `age`.
- The conditional expression `print("You can't vote!") if age < 18 else print("You can vote!")` checks if the value of `age` is less than `18`.
- If the condition `age < 18` evaluates to `True`, the message "You can't vote!" is printed.
- If the condition is `False` (i.e., age is 18 or older), the message "You can vote!" is printed.
- Finally, the message "Done!" is printed after the conditional expression.
