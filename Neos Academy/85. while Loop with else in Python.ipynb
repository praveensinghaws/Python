# 85. while Loop with else in Python


This code searches for the presence of the fruit 'orange' in the list `fruits`. If 'orange' is found, it prints "Orange is available.". Otherwise, it prints "Orange is not available." Here's the code with comments:

```python
# Define a list of fruits
fruits = ['apple', 'banana', 'mango', 'strawberry']

# Get the length of the list
fruits_len = len(fruits)

# Initialize variables
index = 0
fruit_found = False

# Loop through the list of fruits
while index < fruits_len:
    # Check if the current fruit is 'orange'
    if fruits[index] == 'orange':
        # If 'orange' is found, set fruit_found to True and print a message
        fruit_found = True
        print('Orange is available.')
        # Exit the loop since 'orange' is found
        break
    # Move to the next index
    index += 1

# Check if 'orange' is not found after iterating through the entire list
if not fruit_found:
    print('Orange is not available.')
```

In this code:
- The `while` loop iterates over the elements of the `fruits` list.
- Inside the loop, each fruit is checked. If the current fruit is 'orange', `fruit_found` is set to `True`, and the loop is terminated using `break`.
- After the loop, if `fruit_found` is still `False`, it means 'orange' was not found in the list, so it prints "Orange is not available."
