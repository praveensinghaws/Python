# 79.Logical AND with Conditionals in Python


In this code snippet, the variables `age` and `nationality` are assigned values, and an `if` statement is used to check if certain conditions are met. Here's the code with comments:

```python
# Assign the value 20 to the variable age
age = 20

# Assign the string 'Indian' to the variable nationality
nationality = 'Indian'

# Check if age is greater than 18, less than 30, and nationality is 'Indian'
if age > 18 and age < 30 and nationality == 'Indian':
    # If all conditions are true, print the message
    print("You are eligible for the exam.")
```

In this code:
- `age = 20` assigns the value `20` to the variable `age`.
- `nationality = 'Indian'` assigns the string `'Indian'` to the variable `nationality`.
- The `if` statement checks three conditions:
  - `age > 18`: Checks if the age is greater than `18`.
  - `age < 30`: Checks if the age is less than `30`.
  - `nationality == 'Indian'`: Checks if the nationality is `'Indian'`.
- If all conditions are true, the message "You are eligible for the exam." is printed.

In this code snippet, the variables `age` and `nationality` are assigned values, and an `if` statement is used to check if certain conditions are met. If the conditions are not met, the `else` block is executed. Here's the code with comments:

```python
# Assign the value 32 to the variable age
age = 32

# Assign the string 'Indian' to the variable nationality
nationality = 'Indian'

# Check if age is greater than 18, less than 30, and nationality is 'Indian'
if age > 18 and age < 30 and nationality == 'Indian':
    # If all conditions are true, print the message
    print("You are eligible for the exam.")
else:
    # If any condition is false, execute the else block and print the message
    print("You are not eligible for the exam.")
```

In this code:
- `age = 32` assigns the value `32` to the variable `age`.
- `nationality = 'Indian'` assigns the string `'Indian'` to the variable `nationality`.
- The `if` statement checks three conditions:
  - `age > 18`: Checks if the age is greater than `18`.
  - `age < 30`: Checks if the age is less than `30`.
  - `nationality == 'Indian'`: Checks if the nationality is `'Indian'`.
- Since the first condition (`age > 18`) is not true (`32` is not less than `30`), the `else` block is executed, and the message "You are not eligible for the exam." is printed.

```python
age = 20
nationality = 'American'

if age > 18 and age < 30 and nationality == 'Indian':
    print("You are eligible for the exam. Exam fee is Rs.1500.")
elif age > 18 and age < 30 and nationality == 'American':
    print("You are eligible for the exam. Exam fee is $50.")
else:
    print("You are not eligible for the exam.")
```

In this code:
- If the age is between `18` and `30` (inclusive) and the nationality is `'Indian'`, it prints the message "You are eligible for the exam. Exam fee is Rs.1500."
- If the age is between `18` and `30` (inclusive) and the nationality is `'American'`, it prints the message "You are eligible for the exam. Exam fee is $50."
- If the age is not between `18` and `30` (inclusive) or the nationality is neither `'Indian'` nor `'American'`, it prints the message "You are not eligible for the exam."
