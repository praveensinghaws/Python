# 76.The if Statement in Python


The code snippet you provided sets the value of the variable `x` to `10`, and then checks if `x` is less than `50`. If the condition is true, it prints the message "x is less than 50". Here's the code with comments:

```python
# Assign the value 10 to the variable x
x = 10 

# Check if x is less than 50
if x < 50:
    # If the condition is true, print the message
    print("x is less than 50")
```

In this code:
- `x = 10` assigns the value `10` to the variable `x`.
- The `if` statement checks if the value of `x` is less than `50`.
- If the condition `x < 50` evaluates to `True`, the indented block of code under the `if` statement is executed, which contains the `print()` statement that prints the message "x is less than 50".
- Since `x` is indeed less than `50` (with a value of `10`), the message will be printed when the code is executed.

In the provided code snippet, the value of the variable `x` is set to `60`. Then, it checks if `x` is less than `50`. Since `x` is not less than `50`, the condition evaluates to `False`, and the indented block of code under the `if` statement is not executed. However, the code after the `if` statement is executed regardless of the condition. Here's the code with comments:

```python
# Assign the value 60 to the variable x
x = 60

# Check if x is less than 50
if x < 50:
    # This block of code will not be executed
    print("x is less than 50")

# This line is not indented, so it is not part of the if statement block
# It will be executed regardless of the condition
print("End of the program")
```

In this code:
- `x = 60` assigns the value `60` to the variable `x`.
- The `if` statement checks if the value of `x` is less than `50`. Since `x` is not less than `50`, the condition evaluates to `False`.
- The `print("x is less than 50")` statement inside the `if` block will not be executed.
- The `print("End of the program")` statement is not indented, so it is not part of the `if` block. It will be executed regardless of the condition, so it will always print "End of the program" when the code is executed.

The code you provided is functionally equivalent to the previous code snippet, but it's written in a more concise format. Here's the code with comments:

```python
# Assign the value 10 to the variable x
x = 10

# Check if x is less than 50, and if true, print the message
if x < 50: print("x is less than 50")

# Print the message regardless of the condition
print("End of the program")
```

In this code:
- `x = 10` assigns the value `10` to the variable `x`.
- The `if` statement checks if the value of `x` is less than `50`.
- If the condition `x < 50` evaluates to `True`, the message "x is less than 50" will be printed. 
- Regardless of the condition, the message "End of the program" will always be printed after the `if` statement because it's not indented and thus not part of the `if` block.

In this code snippet, there are nested `if` statements. The outer `if` statement checks if `x` is less than `50`, and if true, it further checks if `x` is equal to `10`. If both conditions are true, it prints the messages accordingly. Here's the code with comments:

```python
# Assign the value 10 to the variable x
x = 10

# Check if x is less than 50
if x < 50:
    # If x is less than 50, check if x is equal to 10
    if x == 10:
        # If x is equal to 10, print the message
        print("X is equal to 10")
    # Print the message regardless of whether x is equal to 10
    print("X is less than 50")

# Print the message regardless of whether x is less than 50
print("End of the program")
```

In this code:
- `x = 10` assigns the value `10` to the variable `x`.
- The outer `if` statement checks if the value of `x` is less than `50`.
- If the condition `x < 50` evaluates to `True`, the inner `if` statement checks if the value of `x` is equal to `10`.
- If both conditions are true, the message "X is equal to 10" is printed, and then the message "X is less than 50" is printed.
- Regardless of the conditions, the message "End of the program" is always printed after the `if` statements because it's not indented and thus not part of any `if` block.
