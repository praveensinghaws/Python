# 56. Accessing Elements of a List in Python


```python
# Create a list containing three nested lists
mylist3 = [[1,2,3],[['a','b','c'],[5,6]]]

# Print the entire list
mylist3

# Access the element at index 1 of the first nested list within the second nested list,
# which is the second element ('b') of the innermost nested list
mylist3[1][0][1]
```
