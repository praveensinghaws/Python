# 82.The Basics of while Loop in Python


Your code correctly uses a `while` loop to print the numbers from `1` to `3`. Here's the code with comments:

```python
# Initialize the variable n to 1
n = 1

# Continue looping while n is less than or equal to 3
while n <= 3:
    # Print the current value of n
    print(n)
    # Increment the value of n by 1 for the next iteration
    n += 1
```

In this code:
- `n = 1` initializes the variable `n` to `1`.
- The `while` loop continues to execute as long as the condition `n <= 3` is `True`.
- Inside the loop, the current value of `n` is printed using `print(n)`.
- After printing, the value of `n` is incremented by `1` using `n += 1`. This ensures that the loop progresses towards the termination condition `n <= 3`.
