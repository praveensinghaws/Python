# 61.Input a List using split() Method and Loops in Python


```python
# Read the number of elements as an integer from the user
n = int(input("Enter the number of elements:"))

# Read the numbers as a string from the user and split them into separate elements
numbers = input("Enter the numbers:" ).split()

# Convert each element of the 'numbers' list to an integer using a loop
for i in range(0, n):
    numbers[i] = int(numbers[i])

# Print the list of integers
numbers
```

```python
# Read input as a string, split it into separate numbers, and convert each number to an integer
numbers = [int(x) for x in input("Enter the numbers").split()]
```
