# 64.List Comprehension in Python


```python
# List of names
names = ['Prinka Gandhi', 'Rahul Gandhi', 'Priya', 'Akhilesh Yadav', 'Praveen Singh']

# Empty list to store names containing 'P'
P_Name_List = []

# Iterate over each name in the 'names' list
for name in names:
    # Check if the letter 'P' is in the current name
    if 'P' in name:
        # If 'P' is present in the name, append it to the 'P_Name_List'
        P_Name_List.append(name)

# Print the list of names containing 'P'
P_Name_List
```

In this code:
- We have a list of names stored in the variable `names`.
- We create an empty list called `P_Name_List` to store names containing the letter 'P'.
- We iterate over each name in the `names` list using a `for` loop.
- Inside the loop, we check if the letter 'P' is present in the current name using the `in` operator.
- If 'P' is found in the name, we append the name to the `P_Name_List`.
- Finally, we print the list `P_Name_List`, which contains the filtered names.

```python
# List of animal names
animals = ['lion', 'tiger', 'monkey', 'elephant', 'frog']

# Create a new list 'animals_title_case' using list comprehension
# Convert each animal name to title case using the title() method
animals_title_case = [animal.title() for animal in animals]

# Print the new list 'animals_title_case'
animals_title_case
```
