# 86.for Loop with range() Function in Python


Your code prints the numbers from 0 to 4 using a `for` loop and then prints "Done!". Here's the code with comments:

```python
# Loop over the range of numbers from 0 to 4
for i in range(5):
    # Print the current value of i
    print(i)

# Print 'Done!' after the loop completes
print('Done!')
```

In this code:
- `for i in range(5):` creates a loop that iterates over the numbers from 0 to 4.
- Inside the loop, the current value of `i` is printed using `print(i)`.
- After the loop completes, "Done!" is printed using `print('Done!')`.

Your code prints the numbers from 1 to 5 using a `for` loop and then prints "Done!". Here's the code with comments:

```python
# Loop over the range of numbers from 1 to 5
for i in range(1, 6):
    # Print the current value of i
    print(i)

# Print 'Done!' after the loop completes
print('Done!')
```

In this code:
- `for i in range(1, 6):` creates a loop that iterates over the numbers from 1 to 5 (inclusive).
- Inside the loop, the current value of `i` is printed using `print(i)`.
- After the loop completes, "Done!" is printed using `print('Done!')`.

Your code prints the odd numbers from 1 to 9 using a `for` loop and then prints "Done!". Here's the code with comments:

```python
# Loop over the range of numbers from 1 to 9 with a step of 2
for i in range(1, 10, 2):
    # Print the current value of i
    print(i)

# Print 'Done!' after the loop completes
print('Done!')
```

In this code:
- `for i in range(1, 10, 2):` creates a loop that iterates over the numbers from 1 to 9 with a step of 2 (i.e., it skips even numbers).
- Inside the loop, the current value of `i` (which represents odd numbers) is printed using `print(i)`.
- After the loop completes, "Done!" is printed using `print('Done!')`.
