# 88. for Loop - Reversing a Range of Numbers


Your code prints the numbers from 5 down to 1 in descending order and then prints "Done!". Here's the code with comments:

```python
# Loop over the range of numbers from 5 to 1 with a step of -1
for i in range(5, 0, -1):
    # Print the current value of i
    print(i)

# Print 'Done!' after the loop completes
print("Done!")
```

In this code:
- `for i in range(5, 0, -1):` creates a loop that iterates over the numbers from 5 to 1 in reverse order (i.e., descending).
- Inside the loop, the current value of `i` is printed using `print(i)`.
- After the loop completes, "Done!" is printed using `print("Done!")`.

Your code prints the numbers from 5 down to 1 in descending order using the `reversed()` function with `range()`, and then prints "Done!". Here's the code with comments:

```python
# Loop over the reversed range of numbers from 1 to 5
for i in reversed(range(1, 6, 1)):
    # Print the current value of i
    print(i)

# Print 'Done!' after the loop completes
print("Done!")
```

In this code:
- `reversed(range(1, 6, 1))` creates an iterator that yields the numbers from 5 down to 1 in reverse order.
- The `for` loop iterates over these numbers, and for each iteration, the current value of `i` is printed using `print(i)`.
- After the loop completes, "Done!" is printed using `print("Done!")`.
