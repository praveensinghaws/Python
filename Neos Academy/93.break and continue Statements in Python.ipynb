# 93.break and continue Statements in Python


Here's your code with comments:

```python
# Create a list of numbers from 0 to 99 using the range function
numbers = list(range(0, 100))

# Iterate over each number in the list
for number in numbers:
    # Check if the current number is greater than 50
    if number > 50:
        # If the number is greater than 50, exit the loop
        break
    # If the number is not greater than 50, print it with a space as the end character
    print(number, end=' ')
```

This code iterates over each number from 0 to 99. If a number greater than 50 is encountered, the loop is exited using the `break` statement. Numbers less than or equal to 50 are printed separated by spaces due to the `end=' '` argument in the `print` function.

Your code prompts the user to enter a number continuously until they enter 'q' to quit. Here's the code with comments:

```python
# Run an infinite loop
while True:
    # Prompt the user to enter a number, or 'q' to quit
    num = input("Enter the number (q for quit):")
    
    # Check if the input is 'q'
    if num == 'q':
        # If 'q' is entered, exit the loop
        break
    
    # If the input is not 'q', print the number
    print(num)
```

This code continuously prompts the user to enter a number. If the user inputs 'q', the loop exits, otherwise, it prints the entered number. The loop continues until the user decides to quit by entering 'q'.

Your code is using a `for` loop to iterate over a range of numbers from 0 to 4. Within the loop, it checks if the current number `i` is equal to 2 or 4. If `i` is equal to 2 or 4, it skips the remaining code in the loop and continues to the next iteration using the `continue` statement. Here's the code with comments:

```python
# Iterate over a range of numbers from 0 to 4
for i in range(5):
    # Check if the current number is equal to 2 or 4
    if i == 2 or i == 4:
        # If the current number is 2 or 4, skip the remaining code in the loop
        continue
    # If the current number is not 2 or 4, print it
    print(i)
```

This code will print all numbers from 0 to 4 except for 2 and 4. When `i` is equal to 2 or 4, the `continue` statement is executed, which skips the `print(i)` statement and continues to the next iteration of the loop.

Your code uses a while loop to iterate over numbers from 1 to 10. Within the loop, it increments n by 1 in each iteration. If n is odd (i.e., n % 2 != 0), it skips the remaining code in the loop for that iteration using the continue statement. Here's the code with comments:

```python
# Initialize n to 0
n = 0

# Loop while n is less than or equal to 10
while n <= 10:
    # Increment n by 1 in each iteration
    n += 1
    # Check if n is odd
    if n % 2 != 0:
        # If n is odd, skip the remaining code in the loop for this iteration
        continue
    # If n is even, print it with a space as the end character
    print(n, end=' ')
```

This code will print even numbers from 2 to 10.
