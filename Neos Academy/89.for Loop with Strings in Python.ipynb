# 89.for Loop with Strings in Python


Your code iterates over each character in the string "John" and prints each character separated by a space. After printing all the characters, it prints "Done!". Here's the code with comments:

```python
# Define the string
name = "John"

# Loop over each character in the string
for c in name:
    # Print the current character, followed by a space (end=' ')
    print(c, end=' ')

# Print 'Done!' after printing all characters
print("Done!")
```

In this code:
- The `for` loop iterates over each character (`c`) in the string `name`.
- Inside the loop, each character is printed using `print(c, end=' ')`. The `end=' '` parameter ensures that each character is printed with a space after it, instead of a newline.
- After printing all characters, "Done!" is printed using `print("Done!")`.

Both of your code snippets achieve the same result, which is printing the characters of the string "John" in reverse order, separated by a space, and then printing "Done!".

Here are the explanations for each snippet:

1. Using slicing (`[::-1]`):
```python
name = "John"
for c in name[::-1]:
    print(c, end=' ')
print("Done!")
```
Explanation:
- `name[::-1]` creates a new string containing characters of `name` in reverse order due to the slicing operation.
- The `for` loop iterates over each character (`c`) in the reversed sequence.
- Inside the loop, each character is printed followed by a space using `print(c, end=' ')`.
- After printing all characters, "Done!" is printed using `print("Done!")`.


2. Using `reversed()` function:
```python
name = "John"
for c in reversed(name):
    print(c, end=' ')
print("Done!")
```
Explanation:
- `reversed(name)` creates an iterator that yields characters of the string `name` in reverse order.
- The `for` loop iterates over each character (`c`) in the reversed sequence.
- Inside the loop, each character is printed followed by a space using `print(c, end=' ')`.
- After printing all characters, "Done!" is printed using `print("Done!")`.


Both snippets accomplish the same task, but they use different approaches to reverse the string. The first snippet directly uses the `reversed()` function, while the second snippet uses slicing to create a reversed version of the string.

Your code correctly counts the number of words in the sentence "Hello Rahul how are you" and prints the count. Here's the code with comments:

```python
# Define the sentence
sentence = "Hello Rahul how are you"

# Initialize a counter for the number of words
count = 0

# Split the sentence into words and iterate over each word
for word in sentence.split():
    # Increment the counter for each word encountered
    count += 1

# Print the count of words in the sentence
print(f"There are {count} words in the sentence.")
```

In this code:
- The `sentence.split()` method splits the sentence into words based on whitespace.
- The `for` loop iterates over each word obtained after splitting the sentence.
- Inside the loop, the counter `count` is incremented for each word encountered.
- After the loop completes, the total count of words in the sentence is printed using `print(f"There are {count} words in the sentence.")`.
