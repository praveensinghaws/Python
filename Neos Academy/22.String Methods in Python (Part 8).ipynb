# 22.String Methods in Python (Part 8)


```python
# Finding the last occurrence of 'i' in the string
result1 = "This is a string".rindex('i')
print("Result 1:", result1)

# Finding the first occurrence of 'i' in the string
result2 = "sam2345@gmail.com".find('i')
print("Result 2:", result2)

# Counting the occurrences of 'o' in the string
result3 = "I love Python".count('o')
print("Result 3:", result3)

# Finding the first occurrence of 'p' in the string
result4 = "I love Python".index('p')
print("Result 4:", result4)
```
