# 52.Statements in Python (Part 2)


In the given code, we have both expressions and statements. Here's the breakdown for each line:

1. `x = 10`: 
   - Expression: `10` (the value assigned to `x`)
   - Statement: Assignment statement (`x` is assigned the value `10`)

2. `y = 20`: 
   - Expression: `20` (the value assigned to `y`)
   - Statement: Assignment statement (`y` is assigned the value `20`)

3. `z = x + y * 30`: 
   - Expressions: 
     - `x` (the value of variable `x`)
     - `y` (the value of variable `y`)
     - `30` (literal value)
     - `y * 30` (multiplication expression)
     - `x + y * 30` (addition expression)
   - Statement: Assignment statement (`z` is assigned the result of the expression `x + y * 30`)

4. `p = str(z)`: 
   - Expressions: 
     - `z` (the value of variable `z`)
     - `str(z)` (conversion of `z` to a string)
   - Statement: Assignment statement (`p` is assigned the result of converting `z` to a string)

5. `p = p + str(x)`: 
   - Expressions: 
     - `p` (the value of variable `p`)
     - `str(x)` (conversion of `x` to a string)
     - `p + str(x)` (concatenation of strings)
   - Statement: Assignment statement (`p` is assigned the result of concatenating `p` with the string representation of `x`)

6. `print(p)`: 
   - Expression: 
     - `p` (the value of variable `p`)
   - Statement: Print statement (prints the value of `p` to the console)

So, in summary:
- Each line contains a statement.
- Some lines contain expressions, which are evaluated to produce a value.
- The statements perform assignments, conversions, and printing.
