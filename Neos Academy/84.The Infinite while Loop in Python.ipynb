# 84.The Infinite while Loop in Python


This code will create an infinite loop that continuously prints numbers starting from `100` and decrementing by `1` in each iteration until the program is manually stopped. Here's the code with comments:

```python
# Initialize the variable n to 100
n = 100

# Infinite loop
while True:
    # Print the current value of n
    print(n)
    # Decrement the value of n by 1 for the next iteration
    n -= 1
```

In this code:
- `while True:` creates an infinite loop because the condition `True` is always true.
- Inside the loop, the current value of `n` is printed using `print(n)`.
- The value of `n` is decremented by `1` in each iteration using `n -= 1`. This ensures that the loop continues printing numbers in descending order.

This code creates a loop that continuously prompts the user to enter a line of input until they type `'q'` to quit. Here's the code with comments:

```python
# Infinite loop
while True:
    # Prompt the user to enter a line of input
    line = input("Enter the line (type 'q' to quit): ")
    
    # Check if the user input is 'q'
    if line == 'q':
        # If the user input is 'q', break out of the loop
        break
    
    # Print the user input
    print(line)
```

In this code:
- `while True:` creates an infinite loop because the condition `True` is always true.
- Inside the loop, the user is prompted to enter a line of input using `input("Enter the line (type 'q' to quit): ")`.
- The user input is stored in the variable `line`.
- If the user input is equal to `'q'`, the loop is terminated using `break`.
- Otherwise, the user input is printed using `print(line)`.
