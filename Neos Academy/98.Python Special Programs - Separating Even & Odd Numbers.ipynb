# 98.Python Special Programs - Separating Even & Odd Numbers


```python
# Define a list containing numbers from 1 to 24
li = list(range(1, 25))

# Initialize empty lists to store even and odd numbers
even_numbers = []
odd_numbers = []

# Iterate over each item in the list
for item in li:
    # Check if the item is even
    if item % 2 == 0:
        # If even, append it to the even_numbers list
        even_numbers.append(item)
    else:
        # If odd, append it to the odd_numbers list
        odd_numbers.append(item)
```
