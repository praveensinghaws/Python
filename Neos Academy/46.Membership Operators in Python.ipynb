# 46.Membership Operators in Python


````python
dict = {1: 'Jaspreet', 2: 'Singh'}
'Jaspreet' in dict
````

In the provided code:

```python
dict = {1: 'Jaspreet', 2: 'Singh'}
'Jaspreet' in dict
```

This code creates a dictionary named `dict` with two key-value pairs: `{1: 'Jaspreet', 2: 'Singh'}`. 

The next line `'Jaspreet' in dict` checks whether the string `'Jaspreet'` is a key in the dictionary `dict`. 

However, since the membership operator `in` by default checks for membership of keys in a dictionary, this particular line will check whether `'Jaspreet'` is a key in the dictionary `dict`. 

Since `'Jaspreet'` is not a key in the dictionary (it's a value associated with the key `1`), the output of this line will be `False`. 

Therefore, when you execute this code snippet, the output will be:

```
False
```
