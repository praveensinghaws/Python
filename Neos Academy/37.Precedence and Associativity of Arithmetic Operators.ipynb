# 37.Precedence and Associativity of Arithmetic Operators


To determine the order of execution for the expression `10 + 5 - 8 / 2`, you need to follow the order of operations in mathematics, which is often remembered using the acronym PEMDAS:

1. **P**arentheses
2. **E**xponents (or **O**rders, meaning powers and roots, etc.)
3. **M**ultiplication and **D**ivision (from left to right)
4. **A**ddition and **S**ubtraction (from left to right)

In this case, there are no parentheses or exponents, so you move on to multiplication, division, addition, and subtraction.

Let's break down the expression:

```
10 + 5 - 8 / 2
```

1. Division: `8 / 2` is evaluated first, resulting in `4`.
2. Addition: `10 + 5` is evaluated next, resulting in `15`.
3. Subtraction: The result of the addition (`15`) is then subtracted by the result of the division (`4`), resulting in `11`.

So, the final result of the expression `10 + 5 - 8 / 2` is `11`.
