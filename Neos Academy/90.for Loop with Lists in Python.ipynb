# 90.for Loop with Lists in Python


Your code iterates over each element in the list `cars` and prints each element, followed by "Done!" after printing all elements. Here's the code with comments:

```python
# Define the list of cars
cars = ['Audi', 'BMW', 'Toyota']

# Iterate over each car in the list
for car in cars:
    # Print the current car
    print(car)

# Print "Done!" after printing all cars
print("Done!")
```

In this code:
- The `for` loop iterates over each element (`car`) in the list `cars`.
- Inside the loop, each car is printed using `print(car)`.
- After printing all cars, "Done!" is printed using `print("Done!")`.

Your code iterates over each index of the list `cars`, retrieves the corresponding car using the index, and prints each car. After printing all cars, "Done!" is printed. Here's the code with comments:

```python
# Define the list of cars
cars = ['Audi', 'BMW', 'Toyota']

# Iterate over each index of the list
for i in range(len(cars)):
    # Retrieve the car at index i and print it
    print(cars[i])

# Print "Done!" after printing all cars
print("Done!")
```

In this code:
- The `range(len(cars))` function generates a sequence of indices from 0 to the length of the `cars` list minus 1.
- The `for` loop iterates over each index `i`.
- Inside the loop, `cars[i]` retrieves the car at index `i`, and it is printed using `print(cars[i])`.
- After printing all cars, "Done!" is printed using `print("Done!")`.

Your code uses a list comprehension to iterate over each element in the list `cars`, printing each element directly within the comprehension. After printing all cars, "Done!" is printed. Here's the code with comments:

```python
# Define the list of cars
cars = ['Audi', 'BMW', 'Toyota']

# List comprehension: Iterate over each car in the list and print it
[print(car) for car in cars]

# Print "Done!" after printing all cars
print("Done!")
```

In this code:
- The list comprehension `[print(car) for car in cars]` iterates over each car in the list `cars`.
- For each car, `print(car)` is executed, printing the current car.
- The list comprehension itself doesn't produce a new list; it's primarily used for its side effects (printing in this case).
- After printing all cars, "Done!" is printed using `print("Done!")`.
