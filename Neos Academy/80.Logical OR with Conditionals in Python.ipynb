# 80.Logical OR with Conditionals in Python


Your code is correctly checking if the value of the variable `today` is either `'Saturday'` or `'Sunday'`, indicating a weekend. Here's the code with comments:

```python
# Assign the value 'Sunday' to the variable today
today = "Sunday"

# Check if today is either 'Saturday' or 'Sunday'
if today == 'Saturday' or today == 'Sunday':
    # If today is 'Saturday' or 'Sunday', print the message
    print("It is a weekend!")
```

In this code:
- `today = "Sunday"` assigns the string `'Sunday'` to the variable `today`.
- The `if` statement checks if `today` is equal to `'Saturday'` or `'Sunday'`.
- Since `today` is `'Sunday'`, one of the conditions (`today == 'Sunday'`) is true, so the message "It is a weekend!" is printed.

Your code correctly checks if the value of the variable `today` is either `'Saturday'` or `'Sunday'`. If it's neither of those, it prints that it's a weekday. Here's the code with comments:

```python
# Assign the value 'Monday' to the variable today
today = "Monday"

# Check if today is either 'Saturday' or 'Sunday'
if today == 'Saturday' or today == 'Sunday':
    # If today is 'Saturday' or 'Sunday', print the message
    print("It's is a weekend!")
else:
    # If today is neither 'Saturday' nor 'Sunday', print the message
    print("It's a weekday")
```

In this code:
- `today = "Monday"` assigns the string `'Monday'` to the variable `today`.
- The `if` statement checks if `today` is equal to `'Saturday'` or `'Sunday'`.
- Since `today` is `'Monday'`, neither condition (`today == 'Saturday'` or `today == 'Sunday'`) is true, so the `else` block is executed, and the message "It's a weekday" is printed.

Your code correctly checks the value of the variable `today` to determine the message to print based on whether it's a weekend or a weekday. Here's the code with comments:

```python
# Assign the value 'Tuesday' to the variable today
today = "Tuesday"

# Check if today is either 'Saturday' or 'Sunday'
if today == 'Saturday' or today == 'Sunday':
    # If today is 'Saturday' or 'Sunday', print the message
    print("It's a holiday!")
# Check if today is either 'Monday' or 'Friday'
elif today == 'Monday' or today == 'Friday':
    # If today is 'Monday' or 'Friday', print the message
    print("Work 2 hours extra")
else:
    # If today is neither 'Saturday' nor 'Sunday', nor 'Monday' nor 'Friday', print the message
    print("Normal work hours.")
```

In this code:
- `today = "Tuesday"` assigns the string `'Tuesday'` to the variable `today`.
- The `if` statement checks if `today` is equal to `'Saturday'` or `'Sunday'`. If true, it prints "It's a holiday!".
- The `elif` statement checks if `today` is equal to `'Monday'` or `'Friday'`. If true, it prints "Work 2 hours extra".
- If none of the conditions are met, indicating that it's neither a weekend nor Monday nor Friday, the `else` block is executed, and it prints "Normal work hours."
