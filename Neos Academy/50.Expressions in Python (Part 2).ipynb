Let's break down the expression step by step:

1. `x << 1`: This is a bitwise left shift operation on the value of `x`. Shifting a binary number left by one position is equivalent to multiplying it by 2. So, `x << 1` equals `x * 2`.

2. `x * 3`: This is a multiplication operation, which multiplies the value of `x` by 3.

3. `y + (x<<1) - x * 3`: This expression combines the results of the two previous operations. It adds `y` to the result of `x << 1`, and then subtracts the result of `x * 3`.

Let's calculate it:

Given:
- `x = 10`
- `y = 20`

1. `x << 1` equals `x * 2`, so `x << 1` equals `10 * 2` which is `20`.
2. `x * 3` equals `10 * 3` which is `30`.
3. `z = y + (x<<1) - x * 3` becomes `z = 20 + 20 - 30`, which simplifies to `z = 20 + 20 - 30 = 40 - 30 = 10`.

So, the value of `z` is `10`. 

When you execute this code snippet, the output will be:

```
10
```
