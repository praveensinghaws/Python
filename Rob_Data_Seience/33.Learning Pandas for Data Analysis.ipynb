# Merge Data

## df.merge()

```python
# Group by 'Airline' and 'FlightDate' and calculate the mean of 'DepDelay'
df1 = df.groupby(['Airline', 'FlightDate'])[['DepDelay']].mean().reset_index()

# Group by 'Airline' and 'FlightDate' and calculate the mean of 'ArrDelay'
df2 = df.groupby(['Airline', 'FlightDate'])[['ArrDelay']].mean().reset_index()
```

```python

# Merge df1 and df2 using the default inner join
df_inner = df1.merge(df2)

# Merge df1 and df2 using inner join explicitly
df_inner_explicit = df1.merge(df2, how='inner')

# Merge df1 and df2 using left join
df_left = df1.merge(df2, how='left')

# Merge df1 and df2 using right join
df_right = df1.merge(df2, how='right')

# Merge df1 and df2 using outer join
df_outer = df1.merge(df2, how='outer')
```

# pd.merge()

```python
# Group by 'Airline' and 'FlightDate' and calculate the mean of 'DepDelay'
df1 = df.groupby(['Airline', 'FlightDate'])[['DepDelay']].mean().reset_index()

# Group by 'Airline' and 'FlightDate' and calculate the mean of 'ArrDelay'
df2 = df.groupby(['Airline', 'FlightDate'])[['ArrDelay']].mean().reset_index()
```

```python
# Merge df1 and df2 using the default inner join
pd.merge(df1, df2)

# Merge df1 and df2 on specified columns
pd.merge(df1, df2, on=['Airline', 'FlightDate'])

# Merge df1 and df2 on 'Airline' column with suffixes for disambiguation
pd.merge(df1, df2, on='Airline', suffixes=('_dep', '_arr'))

# Merge df1 and df2 using left_on and right_on parameters
pd.merge(df1, df2, left_on=['Airline'], right_on=['Airline'])
```
