# Measure of Dispersion 
## Range Function
Difference between minimum and maximum value of datasets.

```python
# Range using Default Python
scores = [78,89,92,65,74,80,95,88,70,82]
range = max(scores) - min(scores)
print("Range Using Default Python:",range)
```

```python
# Range using Numpy
import numpy as np
scores = [78,89,92,65,74,80,95,88,70,82]
range = np.ptp(scores)
print("Range using Numpy:",range)
```

```python
# Range using Matplotlib
import matplotlib.pyplot as plt
data = [4,5,6,6,7,7,7,8,8,8,9,9,9,9,10]
plt.hist(data , bins= 6 , edgecolor = 'black' , alpha =0.5 )
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title("Histogram of Data")
plt.axvline(x=min(data), color = 'r' , linestyle = 'dashed' , linewidth =1.5 , label = "Min")
plt.axvline(x=max(data), color = 'g' , linestyle = 'dashed' , linewidth =1.5 , label = "Max")
plt.legend()
plt.show()
```
