# Standard deviation

```python
# Exam 1 Using Python
import statistics as st
scores = [85,90,92,78,88]
std = st.stdev(scores)
print("Standard Deviation is :" ,std)
```

```python
# Exam 2 Using Python
import statistics as st
expense = [600,750,800,700,900]
stdev = st.stdev(expense)
print("Standard Deviation is :" ,stdev)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset
data = np.array([12, 15, 18, 21, 24, 27, 30, 33, 36, 39])

# Calculate standard deviation
std = np.std(data)

# Create a scatter plot
plt.scatter(range(len(data)), data, color='b', label='Data Points')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title(f'Sample Data Scatter Plot\nStandard Deviation: {std:.2f}')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample dataset
data = [12, 15, 18, 21, 24, 27, 30, 33, 36, 39]

# Create a line plot
plt.plot(data, marker='o', linestyle='-', color='b', label='Data Points')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Sample Data Plot')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample dataset
data = [12, 15, 18, 21, 24, 27, 30, 33, 36, 39]
indices = range(len(data))

# Create a bar plot
plt.bar(indices, data, color='b', label='Data Points')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Sample Data Bar Plot')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample dataset
data = [12, 15, 18, 21, 24, 27, 30, 33, 36, 39]

# Create a histogram
plt.hist(data, bins=5, color='b', alpha=0.7, label='Data Distribution')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Sample Data Histogram')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
