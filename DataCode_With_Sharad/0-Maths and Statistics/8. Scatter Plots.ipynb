# Scatter Plots 

A scatter plot is a graphical representation of data points in a two-dimensional space. It's particularly useful for visualizing the relationship between two numerical variables. Each data point in a scatter plot is represented by a dot on the graph, with one variable plotted on the x-axis and the other on the y-axis. The distribution of these dots can help identify patterns, trends, correlations, and outliers in the data.

Here's a brief overview of scatter plots and a couple of examples:

**Key Characteristics:**
- Scatter plots show the relationship between two continuous variables.
- They're used to visualize correlations, trends, and clusters in data.
- The position of each point is determined by its values on the x-axis and y-axis.

**Example 1: Exam Scores vs. Study Hours**

Let's say you have data about students' exam scores and the number of hours they studied. You want to understand if there's a relationship between study hours and exam scores.

```python
import matplotlib.pyplot as plt

study_hours = [2, 3, 5, 1, 4, 6, 7, 8]
exam_scores = [65, 70, 75, 60, 80, 85, 90, 95]

plt.scatter(study_hours, exam_scores)
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.show()
```

In this scatter plot, each point represents a student's study hours and exam score. If most points are trending upward from left to right, it indicates a positive correlation between study hours and exam scores.

**Example 2: Temperature vs Ice Cream Sales**

Suppose you have data about daily temperatures and the number of ice cream cones sold. You want to see if there's any relationship between temperature and ice cream sales.

```python
import matplotlib.pyplot as plt

temperatures = [75, 80, 85, 70, 90, 95, 78, 82]
ice_cream_sales = [120, 140, 160, 100, 180, 200, 130, 150]

plt.scatter(temperatures, ice_cream_sales, color='orange')
plt.title('Temperature vs Ice Cream Sales')
plt.xlabel('Temperature (F)')
plt.ylabel('Ice Cream Sales')
plt.show()
```

In this scatter plot, you can observe if there's a trend where higher temperatures are associated with increased ice cream sales.

Scatter plots are powerful tools for visualizing relationships between two numerical variables. They help you quickly identify patterns and anomalies in your data, making them an essential part of exploratory data analysis.

````python
# Example 1
import matplotlib.pyplot as plt
hours = [2,3,5,1,4,6,7,8]
scores = [65,70,75,60,80,85,90,95]
plt.scatter(hours, scores)
plt.title('Hours Studied vs exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.grid(True)
plt.show()
````
