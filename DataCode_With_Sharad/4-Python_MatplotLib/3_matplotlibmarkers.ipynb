# Matplotlib Markers 

```python
# you can use argument marker to emphasize each point with specified marker.
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([3,8,1,10])
plt.plot(ypoint, marker = 'o')
plt.show()

# We can use '*' insted of 'o'

```

## format strings "fmt" - marker|line|color

```python

from tkinter import Y
import matplotlib.pylab as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, 'o-.r')
plt.show()
```

## marker size - ms

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 15)
plt.show()
```

## marker edge color - mec

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 15, mec = 'r')
plt.show()
```

## # Marker face color - mfc

```python
import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, marker = 'o', ms = 20, mec ='#ff33cc',  mfc = '#00FFFF')
plt.show()
```
