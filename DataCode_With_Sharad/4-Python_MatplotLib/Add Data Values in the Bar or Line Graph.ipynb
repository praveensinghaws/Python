```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt

# Step 2: Create a bar plot
# Sample Data
categories = ['Category 1', 'Category 2', 'Category 3']
values = [20, 35, 45]

# Specify individual colors for each bar
colors = ['c', 'c', 'c']

# Create the bar plot with individual colors
bars = plt.bar(categories, values, color=colors)

# Add data labels to each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 1), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt

# Step 2: Create a bar plot
# Sample Data
categories = ['Category 1', 'Category 2', 'Category 3']
values = [20, 35, 45]

# Specify individual colors for each bar
colors = ['skyblue', 'salmon', 'lightgreen']

# Create the bar plot with individual colors
plt.bar(categories, values, color=colors)

# Add data labels to each bar
for i, value in enumerate(values):
    plt.text(i, value + 0.5, str(value), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt

# Step 2: Create a line plot with markers of different colors
# Sample Data
x_value = [1, 2, 3, 4, 5]
y_value = [10, 25, 13, 32, 20]

# Specify individual colors for each data point
line_color = 'black'
marker_colors = ['blue', 'green', 'orange', 'purple', 'red']

# Create a line plot
plt.plot(x_value, y_value, color=line_color, label='Line')

# Create a scatter plot with markers and different colors for each data point
plt.scatter(x_value, y_value, marker='o', c=marker_colors, label='Markers')

# Add data labels to each point
for x, y in zip(x_value, y_value):
    plt.text(x, y + 0.5, str(y), ha='center', va='bottom')

# Add a title and labels
plt.title('Line Plot with Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
