# Matplotlib Line
#### linestyle or ls - is used to change the style of the plotted line



```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints ,linestyle = 'dotted')
plt.show()
```

## line color - c


```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])
plt.plot(ypoints ,color = 'r')
plt.show()
```

## line width - lw .. width of a line

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linewidth = '10.5' )
plt.show()
```

```python
# example for multiple line
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([3, 8, 1, 10])
ypoints = np.array([6, 2, 7, 11])
plt.plot(xpoints , linewidth = '5.5')
plt.plot(ypoints , linewidth = '5.5')
plt.show()
```
