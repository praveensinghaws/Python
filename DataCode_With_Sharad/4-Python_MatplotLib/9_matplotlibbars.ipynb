# Matplotlib Bars

### Now in this section we will create Vertical Bars bar()

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y)
plt.show()
```

### Now we will create a Horizontal bar

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.barh(x,y)
plt.show()
```

### Color of the bar() and barh()
#### https://matplotlib.org/stable/gallery/color/named_colors.html
#### https://htmlcolorcodes.com/

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y, color="red")
plt.show()
```

### How we can change the bar width

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x,y, width=0.5)
plt.show()
```

### For hoiziontal bar you have to use height instead of width

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.barh(x,y, height=0.5)
plt.show()
```
