```python
import numpy as np
import matplotlib.pyplot as plt

cities = ["Sultanpur", "Lucknow", "Delhi", "Kanpur", "Mumbai"]
population = [10000, 60000, 20000, 40000, 15000]

font1 = {'family': 'serif', 'color': 'k', 'size': 15}
font2 = {'family': 'serif', 'color': 'red', 'size': 20}
colors = np.array([0, 10, 20, 30, 40])  # Adjust the values according to your needs

plt.bar(cities, population, color=plt.cm.plasma(colors / max(colors)), edgecolor='black', width=0.5)
plt.xlabel("City", fontdict=font1)
plt.ylabel("Population", fontdict=font1)
plt.title("City-wise Population Details\n", **font1, weight='bold')  # Remove size and weight from font2
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

cities = ["Sultanpur", "Lucknow", "Delhi", "Kanpur", "Mumbai"]
population = [10000, 60000, 20000, 40000, 15000]

font1 = {'family': 'serif', 'color': 'k', 'size': 15}
font2 = {'family': 'serif', 'color': 'red'}

colors = plt.cm.plasma(np.linspace(0, 1, len(cities)))  # Adjust the colormap based on the number of cities

plt.pie(population, labels=cities, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4, edgecolor='black'))
plt.title("City-wise Population Details\n", **font1, weight='bold')  # Remove size and weight from font2
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mylabels = ["Sultanpur", "Lucknow", "Delhi", "Kanpur", "Mumbai"]
population = [10000, 60000, 20000, 40000, 15000]
myexplode = [0, 0, 0, 0.1, 0]

# Define custom colors for the labels
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'orange']

plt.pie(population, labels=mylabels, startangle=90, explode=myexplode, shadow=True, colors=colors)
plt.legend(title= "Cities")
plt.title("City-wise Population Details", size=15, weight='bold')
plt.show()
```
