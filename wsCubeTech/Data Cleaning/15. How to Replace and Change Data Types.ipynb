```python
# Importing the Pandas library as pd
import pandas as pd

# Importing the Seaborn library as sns
import seaborn as sns

# Importing the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Reading the CSV file "loan.csv" into a DataFrame named df
df = pd.read_csv("loan.csv")

# Displaying the first three rows of the DataFrame
df.head()
```

```python
# Displaying information about the DataFrame
df.info()
```

```python
# Checking for missing values in the DataFrame
df.isnull().sum()
```

```python
# Displaying the counts of unique values in the "Dependents" column
df["Dependents"].value_counts()
```

```python
# Filling missing values in the "Dependents" column with the mode
df["Dependents"].fillna(df["Dependents"].mode()[0], inplace=True)
```

```python
# Checking for missing values in the DataFrame
df.isnull().sum()
```

```python
# Replacing '3+' with '3' in the "Dependents" column
df['Dependents'].replace('3+', '3', inplace=True)
```

```python
# Displaying the counts of unique values in the "Dependents" column
df["Dependents"].value_counts()
```

```python
# Converting the "Dependents" column to the 'int64' data type
df['Dependents'] = df["Dependents"].astype("int64")
```
