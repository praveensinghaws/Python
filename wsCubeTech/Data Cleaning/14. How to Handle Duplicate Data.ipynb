```python
# Importing the Pandas library as pd
import pandas as pd

# Creating a DataFrame with sample data
data = {"Name": ['a', 'b', 'c', 'd', 'a', 'c'],
        "Eng": [8, 7, 5, 8, 8, 5],
        "Hindi": [2, 3, 4, 5, 2, 6]
       }

df = pd.DataFrame(data)
df
```

```python
# Adding a new column "duplicated" to the DataFrame
df["duplicated"] = df.duplicated()

# Displaying the DataFrame with the new column
df
```

```python
# Method 1
# Removing duplicate rows, keeping the first occurrence, and creating a new DataFrame
df_no_duplicates = df.drop_duplicates(keep='first')

# Displaying the new DataFrame without duplicates
df_no_duplicates
```

```python
# Method 2
# Removing duplicate rows, keeping the first occurrence, from the original DataFrame
df.drop_duplicates(keep='first', inplace=True)

# Displaying the modified original DataFrame
df
```

```python
# Method 3
# Removing duplicate rows and creating a new DataFrame
df_no_duplicates = df.drop_duplicates()

# Displaying the new DataFrame without duplicates
df_no_duplicates
```

```python
# Method 4
# Removing duplicate rows from the original DataFrame
df.drop_duplicates(inplace=True)

# Displaying the modified original DataFrame
df
```
