```python
# Importing the Pandas library as pd
import pandas as pd

# Importing the Seaborn library as sns
import seaborn as sns

# Importing the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Reading the CSV file "loan.csv" into a DataFrame named df
df = pd.read_csv("loan.csv")

# Displaying the first three rows of the DataFrame
df.head(3)

```

```python
df.isnull().sum()
```

```python
# Create subplots with 2 rows and 1 column
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Boxplot for the "CoapplicantIncome" column with a different color
sns.boxplot(x="ApplicantIncome", data=df, ax=axes[0], color="skyblue")
axes[0].set_title("Boxplot - ApplicantIncome")

# Distribution plot for the "CoapplicantIncome" column with a different color
sns.histplot(df["ApplicantIncome"], ax=axes[1], kde=True, color="salmon")
axes[1].set_title("Distribution Plot - ApplicantIncome")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
df.describe()
```

```python
# Importing the StandardScaler from scikit-learn
from sklearn.preprocessing import StandardScaler

# Creating an instance of StandardScaler
ss = StandardScaler()

# Fitting the StandardScaler on the "ApplicantIncome" column and transforming the data
df["ApplicantIncome_ss"] = pd.DataFrame(ss.fit_transform(df[["ApplicantIncome"]]), columns=["x"])

# Displaying the first three rows of the DataFrame
df.head(3)

# Displaying summary statistics of the DataFrame
df.describe()
```

```python
# Create subplots with 2 rows and 1 column
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Distribution plot for the "ApplicantIncome" column with a different color
sns.histplot(df["ApplicantIncome"], ax=axes[0], kde=True, color="skyblue")
axes[0].set_title("Before - Distribution Plot - ApplicantIncome")

# Distribution plot for the "ApplicantIncome_ss" column with a different color
sns.histplot(df["ApplicantIncome_ss"], ax=axes[1], kde=True, color="salmon")
axes[1].set_title("After - Distribution Plot - ApplicantIncome_ss")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
