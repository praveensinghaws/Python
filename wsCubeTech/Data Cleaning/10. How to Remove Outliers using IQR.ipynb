```python
# Importing the Pandas library as pd
import pandas as pd

# Importing the Seaborn library as sns
import seaborn as sns

# Importing the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Reading the CSV file "loan.csv" into a DataFrame named df
df = pd.read_csv("loan.csv")

# Displaying the first three rows of the DataFrame
df.head(3)

```

```python
# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot for the "CoapplicantIncome" column
sns.boxplot(x="CoapplicantIncome", data=df, ax=axes[0])
axes[0].set_title("Boxplot - CoapplicantIncome")

# Histogram for the "ApplicantIncome" column
sns.histplot(x="ApplicantIncome", data=df, ax=axes[1])
axes[1].set_title("Histogram - ApplicantIncome")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# Boxplot for the "CoapplicantIncome" column
plt.figure(figsize=(15,5))
sns.boxplot(x="CoapplicantIncome", data=df)
plt.show()
```

```python
# Displaying the shape of the DataFrame
df.shape

# Calculating the first quartile (Q1)
q1 = df["CoapplicantIncome"].quantile(0.25)

# Calculating the third quartile (Q3)
q3 = df["CoapplicantIncome"].quantile(0.75)

# Calculating the Interquartile Range (IQR)
iqr = q3 - q1

# Calculating the minimum and maximum ranges for detecting outliers
min_range = q1 - (1.5 * iqr)
max_range = q3 + (1.5 * iqr)

# Displaying the calculated minimum and maximum ranges
min_range, max_range

# Filtering the DataFrame to include only rows within the calculated range
df_new = df[df["CoapplicantIncome"] <= max_range]

# Displaying the shape of the new DataFrame
df_new.shape
```

```python
# Boxplot for the "CoapplicantIncome" column in the filtered DataFrame
plt.figure(figsize=(15, 5))
sns.boxplot(x="CoapplicantIncome", data=df_new)
plt.show()
```
