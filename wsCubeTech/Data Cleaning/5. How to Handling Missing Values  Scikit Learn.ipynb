```python
# Importing the Pandas library as pd
import pandas as pd

# Importing the Seaborn library as sns
import seaborn as sns

# Importing the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Reading the CSV file "loan.csv" into a DataFrame named df
df = pd.read_csv("loan.csv")

# Displaying the first three rows of the DataFrame
df.head(3)
```

```python
# Displaying the count of missing values in each column of the DataFrame
df.isnull().sum()
```

```python
# Displaying information about the DataFrame, including data types and non-null counts
df.info()
```

```python
# Printing the column names of float64 type in the DataFrame
for i in df.select_dtypes(include='float64').columns:
    print(i)
```

```python
# Importing the SimpleImputer class from scikit-learn for handling missing values
from sklearn.impute import SimpleImputer
```

```python
# Creating an instance of SimpleImputer with the strategy of filling missing values with the mean
si = SimpleImputer(strategy='mean')

# Applying the SimpleImputer to fill missing values in selected columns and obtaining the resulting array
arr = si.fit_transform(df[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']])

```

```python
# Creating a new DataFrame from the array after applying SimpleImputer
df_new = pd.DataFrame(arr, columns=df.select_dtypes(include='float64').columns)

# Displaying the count of missing values in each column of the new DataFrame
df_new.isnull().sum()
```
