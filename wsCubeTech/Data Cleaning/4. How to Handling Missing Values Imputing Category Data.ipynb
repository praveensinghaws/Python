```python
# Importing the Pandas library as pd
import pandas as pd

# Importing the Seaborn library as sns
import seaborn as sns

# Importing the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Reading the CSV file "loan.csv" into a DataFrame named df
df = pd.read_csv("loan.csv")

# Displaying the first three rows of the DataFrame
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Filling missing values using backward fill method along columns
df.fillna(method='bfill')
```

```python
# Filling missing values using backward fill method along rows
df.fillna(method='bfill', axis=1)
```

```python
# Filling missing values using forward fill method along columns
df.fillna(method='ffill')
```

```python
# Filling missing values using forward fill method along rows
df.fillna(method='ffill', axis=1)
```

```python
# Filling missing values in the 'Gender' column with the mode
df['Gender'] = df['Gender'].fillna(df["Gender"].mode()[0])
```

```python
df.head(1)
```

```python
# Counting the number of missing values in columns of object type
df.select_dtypes(include='object').isnull().sum()
```

```python
# Printing the column names of object type in the DataFrame
for i in df.select_dtypes(include='object').columns:
    print(i)
```

```python
# Filling missing values in columns of object type with their respective modes
for i in df.select_dtypes(include='object').columns:
    df[i] = df[i].fillna(df[i].mode()[0])
```
