```python
import pandas as pd

df = pd.DataFrame({"name": ['wscube', 'cow', 'cat', 'dog', 'black']})
```

```python
df
```

```python
# Importing the LabelEncoder from scikit-learn
from sklearn.preprocessing import LabelEncoder

# Creating an instance of LabelEncoder
le = LabelEncoder()

# Applying LabelEncoder to transform the "name" column and creating a new column "en_name"
df["en_name"] = le.fit_transform(df["name"])
```
