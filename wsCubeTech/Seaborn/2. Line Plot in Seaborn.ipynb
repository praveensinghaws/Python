# Line Plot in Seaborn





# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



# Create two lists as sample data
var1 = [1, 2, 3, 4, 5, 6, 7]
var2 = [2, 3, 4, 1, 5, 6, 7]

# Create a DataFrame from the lists
df = pd.DataFrame({"var1": var1, "var2": var2})

# Display the DataFrame
print(df)
```

```python
# Plot 1: Line Plot using Seaborn
sns.lineplot(x='var1', y='var2', data=df)
plt.title('Line Plot using Seaborn')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()
```

```python
# Plot 2: Line Plot using Seaborn with direct variables
sns.lineplot(x=var1, y=var2)
plt.title('Line Plot using Seaborn with Direct Variables')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()
```

```python
# Plot 3: Line Plot using Matplotlib
plt.plot(var1, var2)
plt.title('Line Plot using Matplotlib')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()
```

```python
# Load the penguins dataset from Seaborn
df_sns = sns.load_dataset("penguins").head(50)
df_sns.head(2)
```

```python

# Assuming df_sns is the penguins dataset
sns.set(style="darkgrid")  # Setting a grid style

# Plot 1
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=df_sns, hue='sex', size=50)
plt.title("Line Plot with Size Parameter")
plt.show()

# Plot 2
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=df_sns, hue='sex', style='sex')
plt.title("Line Plot with Different Styles for Each Sex")
plt.show()

# Plot 3
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=df_sns, hue='sex', style='sex', palette='Accent')
plt.title("Line Plot with Custom Palette")
plt.show()

# Plot 4
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=df_sns, hue='sex', style='sex', palette='Accent', markers=['>', 'o'], dashes=False, legend=False)
plt.title("Line Plot with Custom Markers and No Legend")
plt.show()

# Plot 5
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=df_sns, hue='sex', style='sex', palette='Accent', markers=['>', 'o'], dashes=False, legend='full')
plt.title("Line Plot with Custom Markers and Full Legend")
plt.show()

# Plot 6
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=df_sns, hue='sex', style='sex', palette='Accent', markers=['>', 'o'], dashes=False)
plt.grid()
plt.title("Line Plot with Grid")
plt.show()
```
