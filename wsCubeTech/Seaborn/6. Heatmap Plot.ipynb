```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a 4x5 NumPy array
data_array = np.linspace(1, 10, 20).reshape(4, 5)

# Create a heatmap using Seaborn
fig, ax = plt.subplots(figsize=(8, 6))
heatmap = sns.heatmap(data_array, vmin=0, vmax=10, cmap='PuOr', annot=True, annot_kws={'fontsize': 15, 'color': 'c'},
                      linewidths=2, linecolor='k', cbar=False, xticklabels=False, yticklabels=False, ax=ax)

heatmap.set(xlabel='Python', ylabel='Wscube')
sns.set(font_scale=1.5)  # Adjust the font scale
plt.show()
```

```python
# Create a new 2x5 NumPy array
data_new = np.linspace(1, 10, 10).reshape(2, 5)

# Create an array for annotations
annotations = np.array([['a0', 'a1', 'a2', 'a3', 'a4'], ['b0', 'b1', 'b2', 'b3', 'b4']])

# Create a heatmap with annotations
fig, ax = plt.subplots(figsize=(8, 6))
heatmap_new = sns.heatmap(data_new, vmin=0, vmax=10, cmap='Accent', annot=annotations, fmt='s', linewidths=2, ax=ax)

plt.show()
```
