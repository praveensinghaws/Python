```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("penguins.csv")

# Display the first 2 rows of the DataFrame
df.head(2)
```

```python
# Define markers for different sexes
marker = {'MALE': '*', 'FEMALE': 'o'}

# Scatter plot using Seaborn with customizations
sns.scatterplot(
    x="bill_length_mm",
    y="bill_depth_mm",
    data=df,
    hue='sex',      # Color points by 'sex'
    style='sex',    # Differentiate markers by 'sex'
    size='sex',     # Vary marker size by 'sex'
    sizes=(80, 100), # Set size range for markers
    palette='gist_ncar',  # Set color palette
    alpha=0.7,      # Set transparency
    markers=marker  # Set custom markers for each category
)

# Show the plot
plt.show()

```
