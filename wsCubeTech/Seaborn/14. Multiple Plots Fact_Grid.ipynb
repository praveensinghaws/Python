```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
# Scatter plot with FacetGrid, columns='sex', hue='day'
fg = sns.FacetGrid(df, col='sex', hue='day', palette='Accent')
fg.map(plt.scatter, "total_bill", 'tip').add_legend()
plt.show()
```

```python
# Scatter plot with FacetGrid, columns='day', hue='sex'
fg = sns.FacetGrid(df, col='day', hue='sex', palette='summer')
fg.map(plt.scatter, "total_bill", 'tip').add_legend()
plt.show()
```

```python
# Bar plot with FacetGrid, columns='sex', hue='day' (Note: Use 'barplot' for bar plots)
fg = sns.FacetGrid(df, col='sex', hue='day', palette='Accent')
fg.map(sns.barplot, "total_bill", 'tip').add_legend()
plt.show()
```

```python
# Bar plot with FacetGrid, columns='day', hue='sex' (Note: Use 'barplot' for bar plots)
fg = sns.FacetGrid(df, col='day', hue='sex', palette='summer')
fg.map(sns.barplot, "total_bill", 'tip').add_legend()
plt.show()
```
