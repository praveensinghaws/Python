```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv("tips.csv")

# Take a subset of the DataFrame
dfnew = df.head(100)
```

```python
# Plot 1: Strip plot with custom markers
sns.stripplot(x='day', y='total_bill', data=df, hue='sex', palette='rocket_r', jitter=1, size=7, alpha=0.8)
plt.title('Strip Plot with Custom Markers - Full Dataset')
plt.show()
```

```python
# Plot 2: Strip plot with custom markers for a subset of the DataFrame
sns.stripplot(x='day', y='total_bill', data=dfnew, hue='sex', palette='RdPu', linewidth=0.3, edgecolor='c', jitter=2, size=8, alpha=0.7)
plt.title('Strip Plot with Custom Markers - Subset of Data')
plt.show()
```

```python
# Plot 3: Strip plot using a single variable on the x-axis with hue
sns.stripplot(x=df['total_bill'], hue=df['smoker'])
plt.title('Strip Plot with Total Bill and Smoker Hue')
plt.show()
```

```python
# Plot 4: Strip plot using a single variable on the y-axis with hue
sns.stripplot(y=df['total_bill'], hue=df['smoker'])
plt.title('Strip Plot with Total Bill on Y-axis and Smoker Hue')
plt.show()
```
