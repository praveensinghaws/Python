# 5. Bar Plot 

```python
import matplotlib.pyplot as plt

x = ['Python', 'R', 'C++', 'Java']
y = [85, 80, 70, 82]
color = 'skyblue'  

plt.title("Programming Language Popularity", fontsize=15, fontweight='bold')
plt.xlabel("Languages", fontsize=13)
plt.ylabel("Popularity", fontsize=13)

plt.bar(x, y, width=0.5, color=color, align='center', edgecolor='k', linewidth=1, linestyle='--', alpha=0.7, label="Popularity")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['Python', 'R', 'C++', 'Java']
y = [85, 80, 70, 82]
z = [20, 30, 40, 50]

color1 = 'skyblue'  
color2 = 'red' 
color3 = 'blue'
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.title("Programming Language Popularity", fontsize=20, fontweight='bold')
plt.xlabel("Languages", fontsize=13, fontweight='bold')
plt.ylabel("Popularity", fontsize=13, fontweight='bold')

plt.bar(p, y, width, color=color1, align='center', edgecolor='k', linewidth=1, linestyle='--', alpha=0.7, label="Popularity-1")
plt.bar(p1, z, width, color=color2, align='center', edgecolor='k', linewidth=1, linestyle='--', alpha=0.7, label="Popularity-2")

plt.xticks(p + width/2, x, rotation=45, fontsize=12, color=color3)  
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['Python', 'R', 'C++', 'Java']
y = [85, 80, 70, 82]
z = [20, 30, 40, 50]

color1 = 'skyblue'  
color2 = 'red' 
color3 = 'blue'
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.title("Programming Language Popularity", fontsize=20, fontweight='bold')
plt.xlabel("Popularity", fontsize=13, fontweight='bold')
plt.ylabel("Languages", fontsize=13, fontweight='bold')  # Swapped the order of labels

plt.barh(p, y, width, color=color1, align='center', edgecolor='k', linewidth=1, linestyle='--', alpha=0.7, label="Popularity-1")
plt.barh(p1, z, width, color=color2, align='center', edgecolor='k', linewidth=1, linestyle='--', alpha=0.7, label="Popularity-2")

plt.yticks(p + width/2, x, rotation=0, fontsize=12, color=color3)  # Changed to plt.yticks for horizontal bar chart
plt.legend()
plt.show()
```
