```python
import matplotlib.pyplot as plt
import numpy as np

# Data
no = [14, 14, 18, 15, 19, 12, 13, 14, 19, 18, 17, 18, 19, 17, 15, 18, 19, 17, 16, 13, 15, 13, 19, 10,
      12, 15, 14, 15, 18, 18, 18, 14, 15, 13, 15, 13, 18, 11, 13, 18, 17, 10, 18, 10, 18, 14, 11, 16,
      17, 16]

# Adjusted bins for better visualization
plt.hist(no, bins=range(10, 21), edgecolor='black', color='c', bottom=10, density=True, align='mid', histtype='bar', orientation='vertical', rwidth=0.8, label='Distribution')

# Plot title and labels
plt.title("Distribution of Programming Language Scores", fontsize=15, fontweight='bold')
plt.xlabel("Score", fontsize=12, fontweight='bold')
plt.ylabel("Number of Programmers", fontsize=12, fontweight='bold')

# Added x-axis ticks for better clarity
plt.xticks(range(10, 21))

# Threshold line
plt.axvline(15, color='red', linestyle='--', label='Threshold Score')  

# Grid settings
plt.grid(axis='y', linestyle='--', alpha=0.7)  

# Display legend
plt.legend()

# Show the plot
plt.show()
```
