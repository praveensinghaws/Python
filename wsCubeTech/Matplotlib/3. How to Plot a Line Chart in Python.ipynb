# Lec 3. How to Plot a Line Chart in Python



```python
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Scores", fontsize=15)
plt.title("Programming Language Popularity", fontsize=20)

# Use a consistent color for all bars
col = ['c', 'b', 'y', 'k']

# Adjust the width to separate bars
plt.bar(x, y, width=0.5, color=col, align='center', edgecolor='r', linewidth=1, linestyle='--', alpha=0.7, label="Popularity")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]

width = 0.4  # Adjusted width for better separation
p1 = np.arange(len(x))
p2 = [j + width for j in p1]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Scores", fontsize=15)
plt.title("Programming Language Popularity", fontsize=20)

# Adjust the width to separate bars
plt.bar(p1, y, width, color='c', align='center', edgecolor='k', label='Popularity-1')
plt.bar(p2, z, width, color='y', align='center', edgecolor='k', label='Popularity-2')

plt.xticks(p1 + width / 2, x, rotation=20, ha='right')  # Adjusted rotation and alignment for better x-axis labels
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Scores", fontsize=15)
plt.title("Programming Language Popularity", fontsize=20)

# Adjust the width to separate bars
plt.barh(x, y, color='c', align='center', edgecolor='k', label='Popularity-1')
plt.barh(x, z, color='y', align='center', edgecolor='k', label='Popularity-2')

plt.legend(loc='center right')  # Corrected the legend location to 'center right'
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]

bar_width = 0.4
bar_positions_y_pop1 = np.arange(len(x))
bar_positions_y_pop2 = [pos + bar_width for pos in bar_positions_y_pop1]

plt.xlabel("Scores", fontsize=15)
plt.ylabel("Languages", fontsize=15)
plt.title("Programming Language Popularity", fontsize=20)

# Adjust the width to separate bars
plt.barh(bar_positions_y_pop1, y, height=bar_width, color='c', edgecolor='k', label='Popularity-1')
plt.barh(bar_positions_y_pop2, z, height=bar_width, color='y', edgecolor='k', label='Popularity-2')

plt.yticks(bar_positions_y_pop1 + bar_width / 2, x)
plt.legend()
plt.show()
```
