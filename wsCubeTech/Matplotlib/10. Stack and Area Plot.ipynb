```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [2, 3, 2, 4, 2]

# Labels and Colors
lab = ['Area1', 'Area2', 'Area3']
colors = ['r', 'skyblue', 'c']

# Creating a stacked area plot
plt.stackplot(x, area1, area2, area3, labels=lab, colors=colors, baseline='zero')

# Title and Axes labels
plt.title('Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Legend
plt.legend()

# Show the plot
plt.show()
```
