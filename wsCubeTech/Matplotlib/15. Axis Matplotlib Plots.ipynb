```python
import matplotlib.pyplot as plt

# Set the figure size
fig = plt.figure(figsize=(10, 4))

# Plot 1: Line Chart
x = [1, 2, 3, 4]
y = [3, 1, 4, 2]
plt.subplot(1,3,1)
plt.plot(x, y, color='c')
plt.title('Line Chart - One')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0,7)
plt.ylim(0,7)

# Plot 2: Line Chart
x = [1, 2, 3, 4]
y = [3, 1, 4, 2]
plt.subplot(1,3,2)
plt.plot(x, y, color='c')
plt.title('Line Chart - Two')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(x, labels=['Python', 'C', 'C++', 'Java'])  # Set custom labels for X-axis
plt.yticks(y)  # Use default labels for Y-axis

# Plot 3: Line Chart
x = [1, 2, 3, 4]
y = [3, 1, 4, 2]
plt.subplot(1,3,3)
plt.plot(x, y, color='c')
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis([0,7,0,6])  # Set custom axis limits

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
