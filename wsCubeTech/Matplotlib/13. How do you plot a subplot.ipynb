```python
import matplotlib.pyplot as plt


# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Creating subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Subplot 1 - Line chart
axes[0].plot(x, y1)
axes[0].set_title('Line Chart')

# Subplot 2 - Scatter plot
axes[1].scatter(x, y2)
axes[1].set_title('Scatter Plot')

plt.show()


```

```python
import matplotlib.pyplot as plt

# Set a common color palette for all plots
colors = ['r', 'g', 'c', 'skyblue', 'm', 'y']

# Set the figure size
fig = plt.figure(figsize=(12, 6))

# Plot 1: Line Chart
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
plt.subplot(2, 3, 1)
plt.plot(x1, y1, color=colors[0])
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Plot 2: Scatter Plot
x2 = [1, 2, 3, 4, 5]
y2 = [2, 4, 6, 8, 10]
plt.subplot(2, 3, 2)
plt.scatter(x2, y2, color=colors[1])
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Plot 3: Bar Graph
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.subplot(2, 3, 3)
plt.bar(categories, values, color=colors[2])
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')

# Plot 4: Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.subplot(2, 3, 4)
plt.hist(data, bins=5, edgecolor='black', color=colors[3])
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Plot 5: Pie Chart
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 25]
plt.subplot(2, 3, 5)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')

# Plot 6: Stem Plot
# Sample data
x6 = [1, 2, 3, 4, 5]
y6 = [2, 4, 6, 8, 10]
plt.subplot(2, 3, 6)
plt.stem(x6, y6, linefmt=colors[5]+'-', markerfmt=colors[5]+'o', basefmt=colors[5])
plt.title('Stem Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
