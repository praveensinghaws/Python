# Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a boxplot
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])

# Adding labels and title
plt.xlabel('Groups', color= 'r')
plt.ylabel('Values', color= 'r')
plt.title('Boxplot Example')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [10, 20, 30, 40, 60, 70]

# Create a boxplot
plt.boxplot(data, notch=False, vert=True, widths=0.2, labels=['Python'], patch_artist=False, showmeans=True, whis=1.2, sym='r+', boxprops={'color': 'r'}, capprops={'color': 'b'}, whiskerprops={'color': 'c'}, flierprops={'markeredgecolor': 'pink'})

# Adding labels and title
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Boxplot Example')
plt.show()
```
