# Lec 4. Scatterplot



```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]

# Customizations
colors = ['r', 'y', 'g', 'b', 'c', 'pink', 'k']

# Generating sizes for markers based on the 'y' values
sizes = [x * 100 for x in y]

# Plotting
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)
plt.title("Scatter Plot", fontsize=20)

plt.scatter(x, y, color=colors, s=sizes, alpha=0.5, marker='*', edgecolors='pink', linewidths=1)

# Display the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]

# Customizations
colors = [10, 49, 30, 29, 56, 20, 30]

# Generating sizes for markers based on the 'y' values
sizes = [xi * 100 for xi in y]

# Plotting
plt.xlabel("Day", fontsize=15)
plt.ylabel("Number", fontsize=15)
plt.title("Scatter Plot", fontsize=20)

plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.8, marker='o', edgecolors='gray', linewidths=1)

# Adding a colorbar
t = plt.colorbar()
t.set_label("ColorBar", fontsize=15)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the first set of points
day = [1, 2, 3, 4, 5, 6, 7]
no1 = [2, 3, 1, 4, 5, 3, 6]

# Data for the second set of points
no2 = [3, 2, 4, 5, 1, 6, 2]

# Customizations
colors = [10, 49, 30, 29, 56, 20, 30]

# Generating sizes for markers based on the 'no1' values
sizes_no1 = [xi * 100 for xi in no1]

# Generating sizes for markers based on the 'no2' values
sizes_no2 = [xi * 100 for xi in no2]

# Plotting
plt.xlabel("Day", fontsize=15, fontweight='bold')  # Bold labels
plt.ylabel("Number", fontsize=15, fontweight='bold')  # Bold labels
plt.title("Scatter Plot", fontsize=20, fontweight='bold' , c='r')  # Bold title

# Plotting the first set of points
plt.scatter(day, no1, c=colors, s=sizes_no1, cmap='viridis', alpha=0.8, marker='o', edgecolors='pink', linewidths=1, label='Set 1')

# Plotting the second set of points
plt.scatter(day, no2, c='c', s=sizes_no2, alpha=0.8, marker='*', edgecolors='pink', linewidths=1, label='Set 2')

# Adding a colorbar
colorbar = plt.colorbar()
colorbar.set_label("ColorBar", fontsize=15, fontweight='bold' , c= 'b')  # Bold colorbar label

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
