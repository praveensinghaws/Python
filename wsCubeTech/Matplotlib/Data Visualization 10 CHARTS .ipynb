### 3. How to Plot a Line Chart in Python
```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting a line chart
plt.plot(x, y)
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 4. How to Create a Scatterplot in Python Matplotlib?
```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating a scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 5. How to Plot Bar Graph in Matplotlib
```python
# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Creating a bar graph
plt.bar(categories, values)
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### 6. HISTOGRAM CHART IN MATPLOTLIB
```python
# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

# Creating a histogram
plt.hist(data, bins=5, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### 7. How to Create a Pie Chart in Python Matplotlib?
```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 25, 25]

# Creating a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

### 8. Stem Plots with Matplotlib
```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating a stem plot
plt.stem(x, y, linefmt='-', markerfmt='o', basefmt='r')
plt.title('Stem Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 9. How do you Plot a Box and Whisker Plot?
```python
# Sample data
data = [10, 15, 20, 25, 30, 35, 40]

# Creating a box and whisker plot
plt.boxplot(data)
plt.title('Box and Whisker Plot')
plt.ylabel('Values')
plt.show()
```

### 10. How do you Plot Stack and Area Plot
```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 3, 4, 5]

# Creating a stacked area plot
plt.stackplot(x, y1, y2, labels=['Y1', 'Y2'])
plt.title('Stacked Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

### 11. How to Create Step Plot in Python Matplotlib?
```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating a step plot
plt.step(x, y, where='mid')
plt.title('Step Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 12. Fill_Between Plot in Matplotlib Python
```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Creating a fill_between plot
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.5)
plt.title('Fill_Between Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### 13. How do you plot a subplot in Python using Matplotlib
```python
# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Creating subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Subplot 1 - Line chart
axes[0].plot(x, y1)
axes[0].set_title('Line Chart')

# Subplot 2 - Scatter plot
axes[1].scatter(x, y2)
axes[1].set_title('Scatter Plot')

plt.show()
```

### 14. How do I use Savefig in Matplotlib Python

### 15. Axis Matplotlib Plots
```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Creating a plot with specific axis limits
plt.plot(x, y)
plt.title('Plot with Custom Axis Limits')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(1, 6)
plt.ylim(0, 12)
plt.show()
```

### 16. Text in Matplotlib Plots
```python
# Adding text to a plot
plt.text(2, 8, 'Sample Text', fontsize=12, color='red')
plt.title('Plot with Text')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
