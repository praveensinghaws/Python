# Stem Plot

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 5, 8, 3, 6, 9, 4, 7, 1]

# Create a stem plot
plt.stem(x, y, linefmt='b-', markerfmt='bo', basefmt='r-')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stem Plot Example')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 5, 8, 3, 6, 9, 4, 7, 1]

# Create a stem plot with customization
plt.stem(x, y, linefmt='c-', markerfmt='ro', basefmt='b-', bottom=1, label='Python', orientation='vertical')

# Adding labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Stem Plot Example')

plt.show()

```
