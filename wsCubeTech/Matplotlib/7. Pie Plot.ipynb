# Lec 6. Pie Plot

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['skyblue', 'orange', 'lightgreen', 'lightcoral']
explode = (0.1, 0, 0, 0)  # Explode the first slice (Python)

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, shadow=False)

# Title
plt.title("Programming Language Distribution", fontsize=15, fontweight='bold')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['skyblue', 'orange', 'lightgreen', 'lightcoral']
explode = (0.1, 0, 0, 0)  

plt.pie(sizes, labels=labels, explode=explode, colors=colors, autopct='%1.1f%%', startangle=140, shadow=True, radius=1, labeldistance=1.05, textprops={'fontsize': 12, 'color': 'black'}, counterclock=False, wedgeprops={'linewidth': 1, 'width': 0.8, 'edgecolor': 'c'}, rotatelabels=False)

plt.title("Programming Language Distribution", fontsize=15, fontweight='bold')
plt.legend(loc=3)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes1 = [40, 25, 20, 15]
colors = ['skyblue', 'orange', 'lightgreen', 'lightcoral']
sizes2 = [10, 20, 30, 40]

plt.pie(sizes1, labels=labels, radius=1, colors=colors)
plt.pie(sizes2, radius=0.5, colors='w')

plt.title("Programming Language Distribution", fontsize=15, fontweight='bold')

# Specify labels for each set in the legend
plt.legend(labels=['Set 1', 'Set 2'], loc=3)

plt.show()

```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes1 = [40, 25, 20, 15]
colors = ['skyblue', 'orange', 'lightgreen', 'lightcoral']
sizes2 = [10, 20, 30, 40]

plt.pie(sizes1, labels=labels, radius=1, colors=colors)
cr = plt.Circle(xy=(0, 0), radius=0.5, facecolor='w')  # Adjusted radius to make it more centered
plt.title("Programming Language Distribution", fontsize=15, fontweight='bold')
plt.legend(loc=3)
plt.gca().add_artist(cr)
plt.show()
```
