```python
import plotly.graph_objects as go

x = [1, 2, 3, 4]
y = [4, 6, 8, 10]

fig = go.Figure(go.Scatter(
    x=x,
    y=y,
    mode="markers",
    marker=dict(
        size=[40, 60, 80, 100],  # Set marker sizes
        color=['rgb(93,164,214)', 'rgb(255,144,214)', 'rgb(44,160,101)', 'rgb(255,64,54)'],  # Set marker colors
        opacity=[1, 0.8, 0.6, 0.2]  # Set marker opacities
    )
))

fig.show()
```

```python
import plotly.graph_objects as go

x = [1, 2, 3, 4]
y = [4, 6, 8, 10]

fig = go.Figure(go.Scatter(
    x=x,
    y=y,
    text=['Product A', 'Product B', 'Product C', 'Product D'],
    mode="markers",
    marker=dict(
        size=[40, 60, 80, 100],  # Set marker sizes
        color=[10, 12, 14, 16],  # Set marker colors
        opacity=[1, 0.8, 0.6, 0.2],  # Set marker opacities
        showscale=True
    )
))

fig.show()
```
